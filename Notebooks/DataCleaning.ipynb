```python
# ===
# CLEANUP AND COLUMN RENAMING
# ===

# Define rename map
rename_map = {
    # --- Core Identifiers ---
    'Country Code': 'country_code',
    'Country Name': 'country_name',
    'Region ID_x': 'region_id',
    'Region Name_x': 'region_name',
    'Income Group Acronym_x': 'income_group_acronym',
    'Income Group Name_x': 'income_group_name',

    # --- Transaction Types and Years ---
    'Transaction Type ID_x': 'transaction_type_id_total',
    'Transaction Type Name': 'transaction_type_name_total',
    'Fiscal Year_x': 'fiscal_year_total',
    'Transaction Type ID_y': 'transaction_type_id_agency',
    'Transaction Type Name_df2': 'transaction_type_name_agency',
    'Fiscal Year_y': 'fiscal_year_agency',
    'Transaction Type Name_df3': 'transaction_type_name_activity',

    # --- Amounts ---
    'current_amount': 'current_amount_total_aid',
    'constant_amount': 'constant_amount_total_aid',
    'current_amount_df2': 'current_amount_by_agency',
    'constant_amount_df2': 'constant_amount_by_agency',
    'current_amount_df3': 'current_amount_by_activity',
    'constant_amount_df3': 'constant_amount_by_activity',

    # --- Agency Info ---
    'Managing Agency ID': 'managing_agency_id',
    'Managing Agency Acronym': 'managing_agency_acronym',
    'Managing Agency Name': 'managing_agency_name',
    'Managing Sub-agency or Bureau ID': 'sub_agency_id',
    'Managing Sub-agency or Bureau Acronym': 'sub_agency_acronym',
    'Managing Sub-agency or Bureau Name': 'sub_agency_name',

    # --- Funding & Institutional Info ---
    'Funding Account ID': 'funding_account_id',
    'Funding Account Name': 'funding_account_name',
    'Funding Agency ID': 'funding_agency_id',
    'Funding Agency Name': 'funding_agency_name',
    'Funding Agency Acronym': 'funding_agency_acronym',

    # --- Thematic & Sector Classifications ---
    'International Category ID': 'intl_category_id',
    'International Category Name': 'intl_category_name',
    'International Sector Code': 'intl_sector_code',
    'International Sector Name': 'intl_sector_name',
    'International Purpose Code': 'intl_purpose_code',
    'International Purpose Name': 'intl_purpose_name',
    'US Category ID': 'us_category_id',
    'US Category Name': 'us_category_name',
    'US Sector ID': 'us_sector_id',
    'US Sector Name': 'us_sector_name',

    # --- Activity Info ---
    'Activity ID': 'activity_id',
    'Activity Name': 'activity_name',
    'Activity Description': 'activity_description',

    # --- Misc ---
    'Operating Unit': 'operating_unit',
    'OCO Flag': 'oco_flag',
    'Country ID': 'country_id_df3'
}

# Apply renaming
Kenya_Df.rename(columns=rename_map, inplace=True)


# ===
# DROP REDUNDANT / LOW-USE COLUMNS
# ===

cols_to_drop = [
    # Duplicate and non-informative
    'country_code_df2', 'country_code_df3', 'region_name_df3', 'region_id_df3',
    'income_group_id_df3', 'transaction_type_id_total', 'transaction_type_id_agency',

    # Metadata not useful for analysis
    'funding_account_id', 'funding_account_name', 'oco_flag', 'managing_agency_acronym',
    'sub_agency_acronym', 'funding_agency_acronym',

    # Codes
    'us_category_id', 'intl_category_id', 'intl_sector_code', 'intl_purpose_code',

    # Ids
    'sub_agency_id', 'region_id'
]

# Drop columns 
Kenya_Df.drop(columns=cols_to_drop, inplace=True, errors='ignore')

# Sanity check
print(f"Cleaned dataset contains {Kenya_Df.shape[0]} rows and {Kenya_Df.shape[1]} columns")
```

### DATA QUALITY CHECK

```python
# ===
# CHECK FOR DUPLICATES AND MISSINGNESS
# ===

# Check for duplicates
print(f"Duplicates: {Kenya_Df.duplicated().sum()}")

# Check for missing values
missing_summary = Kenya_Df.isnull().sum()
print("\nMissing values by column:", missing_summary[missing_summary > 0])
```

```python
# ===
# FINAL DATA CLEANING & QUALITY CHECKS
# ===

# Drop rows with missing Operating Unit
Kenya_Df.dropna(subset=['operating_unit'], inplace=True)

# Dataset Overview
print("=" * 7)
print(f"Final Cleaned Dataset Summary")
print("=" * 7)
print(f"Rows: {Kenya_Df.shape[0]:,}")
print(f"Columns: {Kenya_Df.shape[1]:,}")
print()

# Data Structure
print("=" * 7)
print("Dataset Info:")
print("=" * 7)
Kenya_Df.info()
print()

# Missing Values Summary
print("=" * 7)
print("Missing Values per Column")
print("=" * 7)

missing_summary = (
    Kenya_Df.isna()
    .sum()
    .reset_index()
    .rename(columns={'index': 'column', 0: 'missing_count'})
)
missing_summary['missing_percent'] = (missing_summary['missing_count'] / len(Kenya_Df)) * 100
missing_summary = missing_summary[missing_summary['missing_count'] > 0]

if missing_summary.empty:
    print("No missing values!")
else:
    display(missing_summary.sort_values(by='missing_count', ascending=False))
```
